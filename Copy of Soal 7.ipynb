```python
list_2 = []

for mean in list_1:
  if mean < 45:
    list_2.append('buruk')
  elif (mean >= 45) & (mean < 80):
    list_2.append('baik')
  elif mean >= 80:
    list_2.append('excellent')

list_2
```
