### Aktivitas 1: Menghitung Peluang dari 2 Kejadian Independen

Saat menghitung peluang untuk beberapa kejadian, hal terpenting yang harus diperhatikan adalah hubungan antar kejadian tersebut, khususnya jika ada irisan antara dua kejadian. Persamaan untuk menghitung peluang kejadian A atau B, jika keduanya memiliki irisan atau sampel point yang sama adalah: 

$P(A \cup B) = P(A) + P(B) - P(A \cap B)$

#### Mutually Exclusive vs Independent untuk Peluang Gabungan (Joint Probability) $P(A \cap B)$

Kejadian A dan B dikatakan *mutually exclusive* jika tidak mungkin kedua kejadian tersebut terjadi bersamaan. Contoh: pelemparan 1 koin, dengan A: kejadian Angka yang terlihat, B: kejadian Gambar yang terlihat. Maka A dan B adalah kejadian yang *mutually exclusive*, tidak ada sample point yang sama antara kedua kejadian tersebut. Untuk kasus ini, maka peluang gabungan $P(A \cap B) = 0$, sehingga

$P(A \cup B) = P(A) + P(B)$

Peristiwa A dan B dikatakan independen jika probabilitas kemunculan B tidak dipengaruhi oleh terjadinya peristiwa A. Misalnya dalam pelemparan koin dua kali; jika A adalah kejadian memperoleh Gambar untuk lemparan pertama dan B adalah kejadian mendapatkan Gambar lagi untuk lemparan kedua, maka jelas hasil lemparan pertama tidak mempengaruhi hasil lemparan yang kedua. Untuk kasus ini, maka peluang gabungan $P(A \cap B) = P(A) \times P(B)$

### Aktivitas 2: Menghitung Peluang Bersyarat (*Conditional Probability*) dari Kejadian Dependen

Misalkan kita ingin menghitung peluang mendapatkan satu kartu Jack dari kartu remi, maka akan kita dapatkan sebesar 
$P(J) = \frac{4}{52} = 7.69 \%$. Namun jika kita ambil kartu tersebut dari tumpukan kartu remi tadi tanpa dikembalikan, lalu hitung peluang mendapatkan kartu Jack dari tumpukan yang sama, maka akan kita dapatkan sebesar $P(\tilde{J}) = \frac{3}{51} = 5.88 \%$. Peluang kartu kedua bergantung pada kartu pertama, dan urutan pengambilan kartu memengaruhi hasil, sehingga kedua kejadian tersebut adalah kejadian yang dependen.

Untuk kasus dependen, maka joint probability diberikan oleh rumus:

<a id='conditional_prob'></a>$P(A,B) = P(A \cap B) = P(A) P(B|A) = P(B) P(A|B)$

Sehingga Conditional Probability diberikan oleh rumus berikut:

$P(B|A) = \frac{P(A \cap B)}{P(A)}$, yang artinya peluang kejadian B setelah diketahui kejadian A terjadi

atau

$P(A|B) = \frac{P(A \cap B)}{P(B)}$, yang artinya peluang kejadian A setelah diketahui kejadian B terjadi

### Aktivitas 3: Hukum Peluang Total (Total Probability Law)

Seringkali kita menemukan permasalahan yang *sample space*-nya atau kemungkinan luarannya dapat dipecah menjadi bagian yang tidak saling tumpang tindih atau tidak ada irisan. Hukum peluang total memungkinkan kita untuk menghitung peluang dalam kondisi tersebut.

Misalnya, dalam tumpukan kartu remi, peluang untuk mendapatkan kartu yang memiliki gambar wajah (*face card*) dapat dihitung dengan mengakumulasi nilai peluang mendapatkan kartu tersebut dalam setiap set-nya (sekop, keriting, hati, atau wajik), yaitu:

$P(face\,card) = P(face\,card \cap sekop) + P(face\,card \cap keriting) + P(face\,card \cap hati) + P(face\,card \cap wajik)$

atau

$P(face\,card) = P(sekop) \cdot P(face\,card | sekop) + P(keriting) \cdot P(face\,card | keriting) \
               + P(hati) \cdot  P(face\,card | hati) + P(wajik) \cdot P(face\,card | wajik)$

### Aktivitas 4: Bayes' Rule

Banyak situasi yang bergantung pada kejadian lain sehingga kita tertarik untuk menghitung peluang dengan mempertimbangkan hubungan tersebut. Namun di sisi lain, banyak permasalahan yang dapat dipelajari dengan mengklasifikasikan elemen dalam ruang sampel tertentu yang tidak memiliki elemen yang sama satu dengan yang lain. Aturan Bayes memungkinkan kita untuk menghitung peluang bersyarat untuk kejadian yang dapat dipartisi menjadi bagian-bagian yang tidak ada irisannya.

Dari [Aktivitas 2](#conditional_prob) kita dapatkan bahwa:

$P(A)\cdot P(B|A) = P(B) \cdot P(A|B)$

maka

$P(B|A) = \frac{P(B) \cdot P(A|B)}{P(A)}$

atau

$P(A|B) = \frac{P(A) \cdot P(A|B)}{P(B)}$

Dengan [hukum peluang total](#total_prob), akan kita dapatkan:

$P(B|A) = \frac{P(B) \cdot P(A|B)}{P(A \cap B_1)+P(A \cap B_2)+\cdots+P(A \cap B_k)}$ untuk `k` kejadian B yang beririsan dengan kejadian A

atau

$P(A|B) = \frac{P(B) \cdot P(A|B)}{P(B \cap A_1)+P(B \cap A_2)+\cdots+P(B \cap A_k)}$ untuk `k` kejadian A yang beririsan dengan kejadian B

### Instruksi

Anda pergi ke dokter untuk memeriksa demam tinggi. Dokter secara acak memilih Anda untuk tes darah untuk penyakit flu burung, yang diduga mempengaruhi 1 dari 9.000 orang di kota Anda. Akurasi dari pengujiannya adalah 99%, artinya peluang positif palsu (*False Positive*) adalah 1%. Peluang negatif palsu (*False Negative*) adalah nol.

Jika hasil tes Anda positif, jawablah pertanyaan-pertanyaan berikut : 

1. Berapa kemungkinan Anda terserang flu burung jika hasil tes Anda positif?

```python
P_Bird_flu = 1/9000
# Peluang tidak terjangkit flu burung
P_no_Bird_flu = 1 - P_Bird_flu
# Peluang hasil tes positive jika diketahui Anda benar-benar terjangkit flu burung
P_Positive_g_Bird_flu = 1
# Peluang hasil tes positive jika diketahui Anda tidak terjangkit flu burung
P_Positive_g_no_Bird_flu = 0.01

# Peluang positif
P_Positive = (P_Bird_flu * P_Positive_g_Bird_flu) + (P_no_Bird_flu * P_Positive_g_no_Bird_flu)

# Bayes' rule untuk P(Bird_flu|Positive)
P_Bird_flu_g_Positive = (P_Bird_flu * P_Positive_g_Bird_flu) / P_Positive
print(P_Bird_flu_g_Positive)
```

expected output: 0.010990218705352238
